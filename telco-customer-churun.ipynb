```python
# 2. Churn rate
print(df["Churn"].value_counts(dropna=False))
```

```python
# 2. Churn rate
churn_rate = round(df["Churn"].value_counts(normalize=True).get("Yes", 0) * 100, 2)
print(f"Churn Rate: {churn_rate}%")
```

```python
# =============================================
# 1. OVERALL CHURN RATE
# =============================================
import pandas as pd
df = pd.read_csv('telco-customer-churn.csv')
total = len(df)
churned = df['Churn'].value_counts().get('Yes', 0)
rate = churned / total
print("Total customers:", total)
print("Churned customers:", churned)
print("Churn rate:", round(rate, 4))
```

```python
# Churn Rate by Contract Type
# Calculate churn rate by contract type
contract_churn = (
    df.groupby('Contract')['Churn']
    .value_counts(normalize=True)
    .unstack()
)

# Rename column if 'Yes' exists
if 'Yes' in contract_churn.columns:
    contract_churn = contract_churn.rename(columns={'Yes': 'Churn Rate'})

print("Churn Rate by Contract Type:")
print(contract_churn.round(4))

print("\nObservation: Month-to-month contracts have significantly higher churn.")
```

```python
payment_churn = (
    df.groupby('PaymentMethod')['Churn']
    .value_counts(normalize=True)
    .unstack()
)

payment_churn = payment_churn.rename(columns={'Yes': 'Churn Rate'})

print(payment_churn.round(4))
```

```python
# =============================================
# CHURN RATE BY MONTHLY CHARGES
# =============================================

print("=== Monthly Charges by Churn ===")
print(df.groupby('churn')['monthlycharges'].mean().round(2))

print("\nInsight: Customers who churn have higher monthly charges on average.")
```

```python
print("=== FINAL SUMMARY & KEY FINDINGS ===")

churn_rate = df['churn'].value_counts(normalize=True).get('Yes', 0) * 100

print(f"• Overall Churn Rate        : {round(churn_rate, 2)}%")
print("• Month-to-month contracts have 3x higher churn")
print("• Electronic check users churn significantly more")
print("• Higher monthly charges strongly correlate with churn")
```
